```python
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data1.csv')
ages = data['Age']
all_dev = data['All_Devs']
py_dev = data['Python']
js_dev = data['JavaScript']
```

```python
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Developers')
plt.plot(ages, py_dev, label='Python Developers')
plt.fill_between(ages, py_dev)
plt.legend()
plt.title("Median Salary (USD) by Ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```

```python
# alpha's value is for the opacity
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Developers')
plt.plot(ages, py_dev, label='Python Developers')
plt.fill_between(ages, py_dev, alpha=0.25)
plt.legend()
plt.title("Median Salary (USD) by Ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```

```python
# Using the overall median
overall_median = 57287
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Developers')
plt.plot(ages, py_dev, label='Python Developers')
plt.fill_between(ages, py_dev, overall_median, alpha=0.25)
plt.legend()
plt.title("Median Salary (USD) by Ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```

```python
#adding where argument
overall_median = 57287
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Developers')
plt.plot(ages, py_dev, label='Python Developers')
plt.fill_between(ages, py_dev, overall_median, where = (py_dev>overall_median), 
                interpolate=True, alpha=0.25)
plt.legend()
plt.title("Median Salary (USD) by Ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```

```python
overall_median = 57287
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Developers')
plt.plot(ages, py_dev, label='Python Developers')
plt.fill_between(ages, py_dev, overall_median, where = (py_dev>overall_median), 
                interpolate=True, alpha=0.25)
plt.fill_between(ages, py_dev, overall_median, where = (py_dev<overall_median), 
                interpolate=True, color='r', alpha=0.25)
plt.legend()
plt.title("Median Salary (USD) by Ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```

```python
#fill between different plots
overall_median = 57287
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Developers')
plt.plot(ages, py_dev, label='Python Developers')
plt.fill_between(ages, all_dev, py_dev,where = (py_dev>all_dev), 
                interpolate=True, alpha=0.25)
plt.fill_between(ages, all_dev, py_dev,where = (py_dev<all_dev), 
                interpolate=True, color='red', alpha=0.25)
plt.legend()
plt.title("Median Salary (USD) by Ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```

```python
#adding labels in fill between method
overall_median = 57287
plt.plot(ages, all_dev, color='#444444', linestyle='--', label='All Developers')
plt.plot(ages, py_dev, label='Python Developers')
plt.fill_between(ages, all_dev, py_dev,where = (py_dev>all_dev), 
                interpolate=True, alpha=0.25, label='Above Average')
plt.fill_between(ages, all_dev, py_dev,where = (py_dev<all_dev), 
                interpolate=True, color='red', alpha=0.25, label='Below Average')
plt.legend()
plt.title("Median Salary (USD) by Ages")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```
