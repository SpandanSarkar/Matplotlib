# counting strings from csv

```python
#importing counters to count how many times the Programming languages have appeared 
from collections import Counter
```

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
print(language_counter)
```

```python
print(language_counter.most_common(15))
#most_common() returns list of touples
```

```python
lang = []
popu = []
for item in language_counter.most_common(15):
    lang.append(item[0])
    popu.append(item[1])
    
print(lang)
print(popu)
```

```python
plt.bar(lang, popu)
plt.title("Most Popular Programming Languages")
plt.xlabel("Programming Languages")
plt.ylabel("Number of people uses")
plt.tight_layout()
plt.show()
```

```python
# making it horizontal for perfect visualisation 
plt.barh(lang, popu)
plt.title("Most Popular Programming Languages")
#plt.ylabel("Programming Languages")
plt.xlabel("Number of people uses")
plt.tight_layout()
plt.show()
```

```python
# reversing the lists
lang.reverse()
popu.reverse()
plt.barh(lang, popu, color='r')
plt.title("Most Popular Programming Languages")
#plt.ylabel("Programming Languages")
plt.xlabel("Number of people uses")
plt.tight_layout()
plt.show()
```

```python
# reading CSV using pandas
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()
    
for response in lang_responses:
    language_counter.update(response.split(';'))

lang = []
popu = []
for item in language_counter.most_common(15):
    lang.append(item[0])
    popu.append(item[1])


plt.barh(lang, popu, color='g')
plt.title("Most Popular Programming Languages")
#plt.ylabel("Programming Languages")
plt.xlabel("Number of people uses")
plt.tight_layout()
plt.show()
```
