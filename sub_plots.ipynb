```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv('sub_plot_data.csv')
```

```python
ages = data['Age']
all_devs = data['All_Devs']
py_devs = data['Python']
js_devs = data['JavaScript']
```

```python
plt.plot(ages, py_devs, label='Python Developers')
plt.plot(ages, js_devs, label='Python Developers')

plt.plot(ages, all_devs, linestyle='--', color='#444444', label='Python Developers')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(ages, py_devs, label='Python Developers')
ax.plot(ages, js_devs, label='Python Developers')

ax.plot(ages, all_devs, linestyle='--', color='#444444', label='Python Developers')

ax.legend()
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```
