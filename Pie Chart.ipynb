```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
slices = [60,40]
lables = ['Sixty','Forty']
```

```python
plt.pie(slices, labels=lables)
plt.title("New Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Seperating with wedeprops
plt.pie(slices, labels=lables, wedgeprops={'edgecolor':'black'})
plt.title("New Pie Chart")
plt.tight_layout()
plt.show()
#matplotlib wedges property 
```

```python
slices1 = [160,40,90,20]
colors = ['#008fd5','#fc4f30','#e5ae37','#6d904f']
lables1 = ['160','40','90','20']
plt.pie(slices1, labels=lables1, colors=colors, wedgeprops={'edgecolor':'black'})
plt.title("New Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Language Popularity
slices2 = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels2 = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
plt.pie(slices2, labels=labels2, wedgeprops={'edgecolor':'black'})
plt.title("Most used Programming Languages")
plt.tight_layout()
plt.show()
# Pie chart is not recomended for this much entities
```

```python
# Language Popularity top 5
# emphasise on some particular slice
# for 3D use shadow
slices3 = [59219, 55466, 47544, 36443, 35917]
labels3 = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.15,0]
plt.pie(slices3, labels=labels3, explode=explode, shadow=True, wedgeprops={'edgecolor':'black'})
plt.title("Most used Programming Languages")
plt.tight_layout()
plt.show()
```

```python
# startagle 
slices3 = [59219, 55466, 47544, 36443, 35917]
labels3 = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.15,0]
plt.pie(slices3, labels=labels3, explode=explode, startangle=90, shadow=True, wedgeprops={'edgecolor':'black'})
plt.title("Most used Programming Languages")
plt.tight_layout()
plt.show()
```

```python
# showing the percentage use attribute "autopct = '%1.1f%%'"
slices3 = [59219, 55466, 47544, 36443, 35917]
labels3 = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.15,0]
plt.pie(slices3, labels=labels3, explode=explode, startangle=90, shadow=True, autopct = '%1.1f%%', wedgeprops={'edgecolor':'black'})
plt.title("Most used Programming Languages")
plt.tight_layout()
plt.show()
```
